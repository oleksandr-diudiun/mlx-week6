# SPICE Metric

Read more about SPICE metric [here](https://ar5iv.labs.arxiv.org/html/1607.08822). Its a semantic matching algorithm based on Stanford's Context Free parser. It only uses caption words and not the images to determine the score between your generated caption and GT.

See image example 234 below to understand
1. SPICE extracts Verb and Noun phrases from your captions(see the example re) "trucks parked"  and "side of the street" from 
your caption and compares it to the ones from GT.  
2. It then computes Precison as the phrases you predicted correctly(True Positives) divided by the ones you predicted but were not in the GT(False Positives).
3. It computes recall by dividing the  True Positives by all the ones in the GT(TP + False negatives).

The score for image 234 are `recall =  0.259` and  `precision = 1`.

And it does Synonyms and lemmatization and also corrects the issues
in BLUE, METEOR which are succeptable to ngram overlap.
